```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
x = [
0.4,
0.5,
0.6,
0.7,
0.8,
]

results = [
62.53088889,
46.29666667,
51.31777778,
36.92666667,
39.70777778,
]

results_ext = [
89.47,
62.77333333,
50.32333333,
55.03,
38.36444444
]
```

```python
fig, ax = plt.subplots(figsize=(5,3))
ax.plot(x, results, label='gods-eye-view')
ax.plot(x, results_ext, label='line-of-sight')
ax.set_xlabel('Cop Speed')
ax.set_ylabel('Average time to capture (seconds)')
ax.legend()
plt.savefig('results.png', dpi=300, bbox_inches='tight')
plt.show()
```
